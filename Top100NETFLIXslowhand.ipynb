```python
import bs4
```

```python
import requests
import pandas as pd
```

```python
name_list = []
title_list = []

data = requests.get('https://www.imdb.com/list/ls063868333/')
soup = bs4.BeautifulSoup(data.text)

for c in soup.find_all("div",{"class" :"lister-item-content"}):
  shirt = c.find("h3",{"class":"lister-item-header"})
  name_list.append(shirt.find('a').text.strip("\n").replace(" ", ""))
  print(name_list)
  #print(name_list)

  # #c.find("div",{"class" :"product-price"})
  # title_list.append(c.find("span",{"class" :"product-price-original"}).text.strip("\n").replace(" ", ""))
  # print(title_list)
  title_list.append(c.find("p",{"class" :""}).text.strip("\n").replace(" ", ""))
  print(title_list)

table = pd.DataFrame([name_list,title_list]).transpose()
table.columns = ['name','title']

table.index +=1

table.to_excel('TOP100NETFLIX.xlsx',engine='openpyxl')
table.to_csv('TOP100NETFLIX.csv', index=False)
```
